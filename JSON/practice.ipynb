Challenge 1: The "Dirty" User List
Scenario: You hit an /users endpoint. Some users are missing the email field, and some have None as their age.
Task: Create a new list called active_emails. Only include emails of users who have an email and are 25 or older.

```python
active_emails = []
for rec in users:
    email = rec.get("email")
    age = rec.get("age")
    if email and age is not None and age >=25:
        active_emails.append(rec)

print(active_emails)
```
