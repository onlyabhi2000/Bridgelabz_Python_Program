A. The Direct Approach: Square Brackets []
This is the most common way.

Usage: data["age"]

Behavior: It is fast and explicit.

Risk: If the key doesn't exist (e.g., data["salary"]), it throws a KeyError and crashes your program immediately.

B. The Safe Approach: The .get() Method
This is used when you aren't 100% sure the key exists.

Usage: data.get("age")

Silent Failure: If the key is missing, it returns None instead of crashing.

Custom Default: You can provide a backup value: data.get("salary", 0). If "salary" isn't there, it returns 0.2

```python
using_get1 = data.get("age")
print(using_get1)
```

```python
using_get2 = data.get("skills")
print(using_get2)

```

```python
# uisng_level3 = data.get("skill_score").get("skill_1")
# print(uisng_level3) --> {'pyhton': 8}
uisng_get3 = data.get("skill_score").get("skill_1").get("zyz",0) # adding a default value if the key's value  is not present
print(uisng_get3)
```

4. Pro Tip: The "Deep" Safety Check
If you are going 3 levels deep, .get() can get messy:
data.get("skill_score").get("skill_1").get("python")

If skill_score is missing, the first .get() returns None. Then you are essentially trying to do None.get("skill_1"), which crashes with an AttributeError.

The Solution: In modern Python (3.10+), we often use "Pattern Matching" or simple if checks:

```python
# uisng_get4 = data.get("skill_score").get("skill_3").get("zyz",0)
# print(uisng_get4) --> AttributeError: 'NoneType' object has no attribute 'get'

## need to add a if condition

if "skill_score" in data and "skill_3" in data["skill_score"]:
    print(data["skill_score"]["skill_1"]["pyhton"])
else:
    print("not present")
```

Final "Level Up" Challenge: The Loop
Since you're learning, the most powerful way to handle nested data isn't by typing [0] or [1], but by using a for loop.

Task: Write a loop that prints the team_name and the language for every team in the list.
