Matplotlib Basics


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
x_axix = np.linspace(0,5,11)
x_axix
```

```python
y_axix = x_axix**2
y_axix
```

```python
plt.plot(x_axix,y_axix)
plt.xlabel("X-axix data")
plt.ylabel("Y-axix data")
plt.title("Linear graph")
```

** SubPlots **

```python
plt.subplot(2,2,1)
plt.plot(x_axix,y_axix)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.plot(x_axix,y_axix)
plt.subplot(2,2,4)
```

** Object Oriented Way **  ---> Full control over the canvas and plot

```python
fig = plt.figure()  ## selecting the canva i.e background
axis1=fig.add_axes([0.1 ,0.1 ,0.8,0.8]) ##(left space , bottom space , %width , %height)
axis1.plot(x_axix,y_axix)
axis1=fig.add_axes([0.3 ,0.3 ,0.4,0.4]) ##(left space , bottom space , %width , %height)
axis1.plot(x_axix,y_axix)
```

```python
plt.scatter(x_axix,y_axix)
```

Types of Plot 

Distribution Plot
We use Distribution plot to visualize the quantitative data   ---> can create a histogram of only one data at a time
1. displot/histplot
2. jointplot
3. pairplot
4. rugplot

```python
import seaborn as sns
```

```python
df =sns.load_dataset('tips')
df
```

```python
df['total_bill'].unique()
```

```python
sns.displot(df['total_bill'], bins=20 , kde = True) ## bins is optional , also bydefault is 20  
## kde is the variation line that is shown --kernal density estimator
```

Multiple plots


```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'])
plt.subplot(1,2,2)
sns.histplot(df['tip'])
```

Disjoint --> can compare diffefent data together


```python
sns.jointplot(x='total_bill' , y ='tip', data=df,kind = )
```

```python
sns.pairplot(df , hue='sex')
```
