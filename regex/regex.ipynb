1. match() -->  Checks if the start of the string matches the pattern.

```python
string = 'Hello world'
regex = 'Hello'
result = re.match(regex , string)
print(result)
```

2. re.search()  --> Finds the first match anywhere in the string.

```python
string = 'Hi! , this is anna , an AI based code suggestion application'
regex = 'code'
result = re.search(regex , string)
print(result)
```

3. re.findall() --> Returns a list of all matches.

```python
string = 'Hi! , this is anna , an AI based code suggestion application this is my version 2 and next will get the update for the version 3'
regex = '\d'
result = re.findall(regex , string)
print(result)
```

4. re.finditer()  --> Returns iterators of match objects (with details).



```python
string = 'Hi! , this is anna , an AI based code suggestion application this is my version 2 and next will get the update for the version 3'
regex = '\d'
for match in re.finditer(regex , string):
    print(match.group() , match.start() , match.end())
```

| Code Part                    | Explanation                                |
| ---------------------------- | ------------------------------------------ |
| `regex = r'\d'`              | Matches any single digit 0–9               |
| `re.finditer(regex, string)` | Finds *all* digit matches as Match objects |
| `match.group()`              | The matched digit as string                |
| `match.start()`              | Start index in original string             |
| `match.end()`                | End index (just after the digit)           |


5. re.sub()  -->  Replaces matched text.

```python
string = 'i like cats because cats are cute.'
new_string = re.sub(r'cats' , 'dogs' , string , re.IGNORECASE)
print(new_string)
```

```python
import re

text = "I like cats because Cats are cute."

# Define replacements
replacements = {
    'cats': 'dogs',
    'cute': 'adorable'
}

# Create the pattern
pattern = r'|'.join(re.escape(word) for word in replacements.keys())

# Replace function
def replace(match):
    return replacements[match.group(0).lower()]

# Perform substitution
new_text = re.sub(pattern, replace, text, flags=re.IGNORECASE)

print(new_text)
```

6. re.split()  --> Splits text by the pattern.

```python
string = 'i am learning regex'
result = re.split(r'\s' , string)
print(result)
```

```python
string = "apple,banana;orange grape"
result = re.split(r'[,; ]+' , string)
print(result)
```
