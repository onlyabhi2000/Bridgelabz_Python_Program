
q1. Extract all numbers

Given a string:
"Order 123 was placed on 2022-05-30 and costs 49.99 USD"
Extract all numbers (integers and decimals).


```python

import re
string = 'Order 123 was placed on 2022-05-30 and costs 49.99 USD'
regex = r'\d+(?:\.\d+)?'

result = re.findall(regex,string)
print(result)
```

q3.  Validate email

Write a regex that matches valid email addresses.
Test with:
"user@example.com", "bad@.com", "another.user@domain.org"

email = user@example.com
regex = '\w+\@\w+\'

```python
import re

emails = [
    "user@example.com",
    "bad@.com",
    "another.user@domain.org"
]

# Regex pattern
pattern = r'^[\w\.-]+@[\w-]+\.[a-zA-Z]{2,3}$'

for email in emails:
    if re.match(pattern, email):
        print(f"{email} is valid ")
    else:
        print(f"{email} is invalid ")
```

q5. Replace all digits with X

"My PIN is 1234 and OTP is 5678"
Result: "My PIN is XXXX and OTP is XXXX"

```python
string = 'My Pin is 1234 and otp is 5678'
new_string = re.sub(r'[\d]' , 'X' , string)
print(new_string)
```

q6.  Extract dates in various formats

From text:
"Meeting on 2023-06-21, backup on 21/06/2023"
Extract both dates.

```python
string = 'Meeting on 2023-06-21 , backup on 21/06/2023'
#For YYYY-MM-DD: \d{4}-\d{2}-\d{2}
# For DD/MM/YYYY: \d{2}/\d{2}/\d{4}

pattern = r'\d{4}-\d{2}-\d{2}|\d{2}/\d{2}/\d{4}'

result = re.findall(pattern , string)
print(result)
```

q9. Reformat dates

Convert dates from "DD/MM/YYYY" to "YYYY-MM-DD"
Example: "21/06/2023" → "2023-06-21"


```python
string = 'Convert dates from "21/06/2023" to "2023-06-21"'
# Correct pattern with capturing groups
pattern = r'(\d{2})/(\d{2})/(\d{4})'
replacement = r'\3-\2-\1'

result = re.sub( pattern , replacement , string )
print(result)
```

q10. Extract URLs from text

"Visit https://example.com or http://test.org."
Get: ['https://example.com', 'http://test.org']

```python
string = "Visit https://example.com or http://test.org."
pattern = r'https?://\S+'

matches = re.findall(pattern , string)
result = [url.rstrip('.,!?') for url in matches]
print(result)
```
